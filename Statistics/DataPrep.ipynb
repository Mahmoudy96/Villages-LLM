### read file

```python
import pandas as pd

file_path = 'Data/VillageStatistics.xls'
sheet_name = 'VillageStatistics11'  # Adjust based on the actual sheet name
```

```python
header_rows = pd.read_excel(file_path, sheet_name=sheet_name, header=None, skiprows=3, nrows=3)

# Combine the three rows into a single header row
# Use a space or underscore to join the text in each column
combined_headers = header_rows.apply(lambda x: ' '.join(x.dropna().astype(str)), axis=0)

# Read the data starting from the 6th row (after the headers)
df = pd.read_excel(file_path, sheet_name=sheet_name, header=None, skiprows=6)

# Set the combined headers as the column names
df.columns = combined_headers

# Clean up column names (optional: remove extra spaces, special characters, etc.)
df.columns = df.columns.str.strip().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')

# combine Village and District columns to avoid duplicate values in village
df['Village'] = df['Village'] + ' (' + df['District'].str.replace(' ','') + ')'

df = df.fillna(0)

# Save the cleaned data to a new Excel file
output_file = 'Data/cleaned_Village_data.xlsx'
df.to_excel(output_file, index=False)
```
